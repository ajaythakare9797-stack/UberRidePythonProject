```python
# ðŸš• Uber Ride Analysis Project

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")
np.random.seed(42)

# Generate synthetic data
num_records = 1000
locations = ['Manhattan', 'Brooklyn', 'Queens', 'Bronx', 'Staten Island']
ride_types = ['UberX', 'UberXL', 'UberPool', 'UberBlack']
times = pd.date_range(start='2023-01-01', end='2023-12-31', periods=num_records)

uber_data = pd.DataFrame({
    'Ride ID': np.arange(10000, 10000 + num_records),
    'Pickup Location': np.random.choice(locations, num_records),
    'Dropoff Location': np.random.choice(locations, num_records),
    'Ride Type': np.random.choice(ride_types, num_records),
    'Distance (km)': np.round(np.random.exponential(scale=5, size=num_records), 2),
    'Fare Amount ($)': np.round(np.random.normal(loc=20, scale=8, size=num_records), 2),
    'Tip Amount ($)': np.round(np.random.exponential(scale=2, size=num_records), 2),
    'Ride Timestamp': times
})

uber_data['Hour'] = uber_data['Ride Timestamp'].dt.hour
uber_data['Day of Week'] = uber_data['Ride Timestamp'].dt.day_name()

# Save CSV (optional)
uber_data.to_csv("Uber_Ride_Data.csv", index=False)
print("âœ… Dataset created!")

# Example Plot: Average Fare by Ride Type
plt.figure(figsize=(8,5))
sns.barplot(x='Ride Type', y='Fare Amount ($)', data=uber_data)
plt.title("Average Fare by Ride Type")
plt.show()

#Number of Rides by Day of the Week

plt.figure(figsize=(8, 5))
sns.countplot(x='Day of Week', data=uber_data, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title("Number of Rides by Day of the Week")
plt.xticks(rotation=45)
plt.show()

#Distance vs Fare Scatter Plot

plt.figure(figsize=(8, 5))
sns.scatterplot(x='Distance (km)', y='Fare Amount ($)', data=uber_data, hue='Ride Type')
plt.title("Distance vs Fare Amount")
plt.show()


#Average Tip by Ride Type

plt.figure(figsize=(8, 5))
sns.barplot(
    x='Ride Type',
    y='Tip Amount ($)',
    data=uber_data,
    hue='Ride Type',
    palette='Blues_d',
    legend=False
)
plt.title("Average Tip by Ride Type")
plt.show()


#Distribution of Fare Amounts

plt.figure(figsize=(8, 5))
sns.histplot(uber_data['Fare Amount ($)'], bins=30, kde=True, color='green')
plt.title("Distribution of Fare Amounts")
plt.xlabel("Fare ($)")
plt.ylabel("Frequency")
plt.show()







```
